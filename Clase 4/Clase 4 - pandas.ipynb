```python
Autos = pd.DataFrame({'modelo':modelos, 'Peso':peso, 'Precio':precios}) #Diccionario

Autos
```

```python
precios
```

```python
#series
Autos_peso = pd.Series(peso, index=modelos)
Autos_peso
```

```python
#Crear Dataframe
Autos = pd.DataFrame({'Peso':Autos_peso, 'Precio':Autos_precio}) #Diccionario
Autos
```
